```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\King alagbe\Videos\CV\DATASET\poses_csv.csv")
```

```python
X = df
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_trans = sc.fit_transform(X.drop(["JPG" , "Class"], axis = 1))
train_trans
```

```python
X = train_trans
Y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,  y_test  = train_test_split(X, Y , test_size = 0.2, random_state = 900)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier(base_score=0.2, booster='gbtree', colsample_bylevel=1,
       colsample_bytree=1, gamma=0, learning_rate=0.1, max_delta_step=2,
       max_depth=3, min_child_weight=3, missing=None, n_estimators=109,
       n_jobs=4, nthread=None, objective='binary:logistic', random_state=4,
       reg_alpha=0, reg_lambda=1, scale_pos_weight=23.9, seed=None,
       silent=True, subsample=1)
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import seaborn as sns
from sklearn.metrics import classification_report , confusion_matrix
cm = np.array(confusion_matrix(y_test, y_pred))
print("Confusion matrix:\n" , cm)
sns.heatmap(cm, annot =True)
```

```python
cr = classification_report(y_test, y_pred)
print("Classification report : \n" , cr)
```
