```python
#import data
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("poses_csv.csv")
```

```python
df.head()
```

```python
#check for statistical details
df.describe()
```

```python
X = df
```

```python
#perform preprocessing on the features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_trans = sc.fit_transform(X.drop(["JPG" , "Class"], axis = 1))
train_trans
```

```python
# convert the classes to integers
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['Class'] = labelencoder.fit_transform(df['Class'])
```

```python
X = train_trans
Y = df['Class']
```

```python
#split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,  y_test  = train_test_split(X, Y , test_size = 0.2, random_state = 900)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import lightgbm as lgb
litgb = lgb.LGBMClassifier(silent=False, learning_rate =  0.1, max_depth = 4, n_estimators = 200, num_leaves = 150)
litgb.fit(X_train, y_train)
```

```python
# Check how well the model Classified the data
from sklearn.metrics import classification_report , confusion_matrix
import seaborn as sns
litgb_pred = litgb.predict(X_test)
con_matrix = np.array(confusion_matrix(y_test, litgb_pred))
print("Confusion matrix:\n" , con_matrix)
sns.heatmap(con_matrix, annot =True)
class_report = classification_report(y_test, litgb_pred)
print("Classification report : \n" , class_report)
```
